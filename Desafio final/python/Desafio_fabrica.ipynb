```python
import pandas as pd
import matplotlib.pyplot as plt



dados = pd.read_csv("/content/dados_rh.csv")
```

```python
dados.shape

#o dataframe possui 1470 linhas e 35 colunas
```

```python
display(dados.head())
```

```python
dados.isnull().sum()

#Nao contem dados nulos
```

```python
tipos_de_dados = dados.dtypes.value_counts()
print(tipos_de_dados)

#esta presente dados do tipo int e object
```

```python
plt.figure(figsize=(8, 5))
dados["Department"].value_counts().plot(kind="bar", color=["blue", "green", "red"])
plt.title("Quantidade de funcionários por Setor")
plt.xlabel("setor")
plt.ylabel("Quantidade de funcionarios")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
dados["Education"].value_counts().sort_index().plot(kind="line", marker="o", color="purple", linestyle="-")
plt.title("quantidade de funcionarios por nivel de escolaridade")
plt.xlabel("Nível de Escolaridade")
plt.ylabel("quantidade de funcionarios")
plt.xticks(range(dados["Education"].min(), dados["Education"].max() + 1))
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
dados["Attrition"].value_counts().plot(kind="pie", autopct="%1.1f%%", colors=["lightblue", "orange"], startangle=90)
plt.title("proporção de funcionarios com ou sem atrito na empresa")
plt.ylabel("")
plt.show()
```
